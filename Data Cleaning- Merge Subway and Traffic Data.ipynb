```python
#modified from: 
#https://stackoverflow.com/questions/7274267/print-all-day-dates-between-two-dates
from datetime import date, timedelta
def dates(startyear, startmon, startday, endyear, endmon, endday): #in (year, month, day) format
    start_date = date(startyear, startmon, startday) 
    end_date = date(endyear, endmon, endday)    # perhaps date.now()

    delta = end_date - start_date   # returns timedelta
    days = []
    for i in range(delta.days + 1):
        day = start_date + timedelta(days=i)
        days.append(str(day))
    return days
```

```python
#write a function to average over each day in the month for the street of interest
def daily_average_speed(df, startyear, startmon, startday, endyear, endmon, endday):
    daily_avg_df = pd.DataFrame()
    days = dates(startyear, startmon, startday, endyear, endmon, endday)
    means = []
    streets = []
    directions = []
    segment_ids = []
    weekday = []
    for day in days:
        #get all valid data from a specific day
        one_day_df = df[df['time'].str.contains(day)]
        avg_speed_one_day = np.mean(one_day_df['speed'])
        means.append(avg_speed_one_day)
        streets.append(np.unique(one_day_df['street']))
        print(np.unique(one_day_df['street']))
        directions.append(np.unique(one_day_df['direction']))
        segment_ids.append(np.unique(one_day_df['segment_id']))
        weekday.append(np.unique(one_day_df['day_of_week']))
    daily_avg_df['segment_id'] = segment_ids
    daily_avg_df['day'] = days
    daily_avg_df['avg_daily_speed'] = means
    daily_avg_df['street'] = streets
    daily_avg_df['direction'] = directions
    daily_avg_df['day_of_week'] = weekday
    return daily_avg_df

        
        
   
```
