```python
#modified from: 
#https://stackoverflow.com/questions/7274267/print-all-day-dates-between-two-dates
from datetime import date, timedelta
def dates(startyear, startmon, startday, endyear, endmon, endday): #in (year, month, day) format
    start_date = date(startyear, startmon, startday) 
    end_date = date(endyear, endmon, endday)    # perhaps date.now()

    delta = end_date - start_date   # returns timedelta
    days = []
    for i in range(delta.days + 1):
        day = start_date + timedelta(days=i)
        days.append(str(day))
    return days
```

```python
#write a function to average over each day in the month for the street of interest
def daily_average_speed(df, startyear, startmon, startday, endyear, endmon, endday):
    daily_avg_df = pd.DataFrame(columns=['segment_id', 'day', 'avg_daily_speed', 'street', 'direction', 'day_of_week'])

    days = dates(startyear, startmon, startday, endyear, endmon, endday)
    for day in days:
        #get all valid data from a specific day
        one_day_df = df[df['time'].str.contains(day)]
        if (np.unique(one_day_df['segment_id']) != nan):
            segment_id = np.unique(one_day_df['segment_id']).item(0)
            avg_speed_one_day = np.mean(one_day_df['speed'])
            street = np.unique(one_day_df['street']).item(0)
            direction = np.unique(one_day_df['direction']).item(0)
            day_of_week = np.unique(one_day_df['day_of_week']).item(0)
            print(segment_id, day, avg_speed_one_day, street, direction, day_of_week)
            daily_avg_df.append(pd.Series([segment_id, day, avg_speed_one_day, street, direction, day_of_week]), ignore_index=True)

    return daily_avg_df
   
```

```python
daily_avg_traffic_df_closed = daily_average_speed(traffic_df_closed_clean,2021,5,16, 2021,6,16)
daily_avg_traffic_df_open = daily_average_speed(traffic_df_open_clean,2021,4,16, 2021,5,16)

print(daily_avg_traffic_df_closed)
```
