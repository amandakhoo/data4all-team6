## Mark certain days for maintenance
dates selected from scanning CTA & Chicago news websites and Twitter feeds

```python
# days of station closure
closures = {'start_date' : ['2020-08-10', '2020-01-24', '2020-01-31', '2020-02-07', '2020-02-14', '2020-02-21',
                          '2020-02-29', '2019-12-13', '2021-06-20', '2018-08-10', '2019-09-27', '2019-07-12',
                          '2019-08-28'],
          'end_date' : ['2020-08-11', '2020-01-27', '2020-02-03', '2020-02-10', '2020-02-17', '2020-02-24',
                       '2020-03-01', '2019-12-16', '2021-06-21', '2018-08-13', '2019-10-07', '2019-07-15',
                       '2019-08-29'],
          'description' : ['Fullerton-Cermak, Ashland-lakefront', 'Granville, Thorndale, Bryn Mawr',
                          'Granville, Thorndale, Bryn Mawr', 'Granville, Thorndale, Bryn Mawr, Chicago', 
                          'Granville, Thorndale, Bryn Mawr', 'Granville, Thorndale, Bryn Mawr, Grand',
                          'Adams/Wabash, Clark/Lake, Washington/Wabash, State/Lake', 'Northbound Red Line Wilson-Howard',
                          'Brown Line Belmont-Loop', 'Belmont', "O'Hare", 'Morgan', 'Belmont']}

dates = pd.to_datetime(df['date']) # convert dates to datetime for ease of comparison

# create a binary variable, to track maintenance through each date in dates
maintenance = pd.Series(np.zeros(len(dates)), name='maintenance')

# check if a date falls within each maintenance period, then combine list of booleans to represent all dates in "closures"
for t in [(dates >= start) & (dates < end) for start, end in zip(closures['start_date'], closures['end_date'])]:
    maintenance = pd.Series(np.logical_or(maintenance, t), name='maintenance')

# combine dates and maintenance booleans to make one dataframe
maint_df = pd.concat([dates, maintenance], axis=1)
```

## Principal component analysis
[Pandas docs on dealing with missing values -- interpolation](https://pandas.pydata.org/docs/user_guide/missing_data.html#interpolation) \
[sklearn on imputing missing variables](https://scikit-learn.org/stable/modules/impute.html#estimators-that-handle-nan-values) \
[sklearn on PCA](https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html)

```python
from sklearn.preprocessing import StandardScaler

# remove NaN values by interpolating -- still had issues w/ PCA, so filled too
clean_df = df.interpolate(axis=0, limit_direction='both').fillna(-1)
# separate X (time) and Y (features)
X = clean_df.iloc[:,0] # date column
Y = clean_df.iloc[:,1:] # everything after date column

# convert date to datetime object
X = pd.to_datetime(X)

# normalize feature space
Y = Y / Y.max()
y = StandardScaler().fit_transform(Y)
```

```python
from sklearn.decomposition import PCA

# initialize PCA
pca = PCA()
Y_r = pca.fit_transform(X=Y) 
y_r = pca.fit_transform(X=y)
Ym_r = pca.fit_transform(X=pd.concat([Y, maint_df['maintenance']], axis=1))

df_pca = pd.DataFrame(Ym_r) # store as dataframe to view 

print(df_pca.shape)
df_pca.round(2).head()
```

```python
# # principal weight components
# df_pca_loadings = pd.DataFrame(pca.components_)

# # mean centered data
# Y_mean = Y - Y.mean()

# print contribution of the first 10 PC's to total variance 
#        (sum across all PC's = 1)
print(pca.explained_variance_ratio_.round(2)[:10])
```

```python
# define seasons by time of year, may be a day off sometimes due to leap years
winter_mask = (X.dt.strftime('%m-%d') >= '12-21') | ((X.dt.strftime('%m-%d') >= '01-01') & (X.dt.strftime('%m-%d') < '03-20'))
spring_mask = (X.dt.strftime('%m-%d') >= '03-20') & (X.dt.strftime('%m-%d') < '06-21')
summer_mask = (X.dt.strftime('%m-%d') >= '06-21') & (X.dt.strftime('%m-%d') < '09-22')
fall_mask = (X.dt.strftime('%m-%d') >= '09-22') & (X.dt.strftime('%m-%d') < '12-21')

# cluster by seasons
fig, ax = plt.subplots(dpi=100)
plt.scatter(Y_r[winter_mask,0], Y_r[winter_mask,1], alpha=0.5, label='winter')
plt.scatter(Y_r[spring_mask,0], Y_r[spring_mask,1], alpha=0.5, label='spring')
plt.scatter(Y_r[summer_mask,0], Y_r[summer_mask,1], alpha=0.5, label='summer')
plt.scatter(Y_r[fall_mask,0], Y_r[fall_mask,1], alpha=0.5, label='fall')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend()
plt.title('PCA of Chicago traffic congestion & subway ridership datasets', fontsize=12);
```

## clustering after processing to match dates as feature

```python
# remove NaN's
c = df.dropna(axis=1)

# extract columns to be target and features
x = c.iloc[:,0] # date
y = c.iloc[:,1:]

# turn date into datetime object
x = pd.to_datetime(x)

# standardize features by normalizing y
y = StandardScaler().fit_transform(y)
```

```python
# make masks of date-related details
seasons = pd.Series(np.zeros(len(x)), name='seasons')
seasons[(x.dt.strftime('%m-%d') >= '12-21') | ((x.dt.strftime('%m-%d') >= '01-01') & (x.dt.strftime('%m-%d') < '03-20'))] = 'winter'
seasons[(x.dt.strftime('%m-%d') >= '03-20') & (x.dt.strftime('%m-%d') < '06-21')] = 'spring'
seasons[(x.dt.strftime('%m-%d') >= '06-21') & (x.dt.strftime('%m-%d') < '09-22')] = 'summer'
seasons[(x.dt.strftime('%m-%d') >= '09-22') & (x.dt.strftime('%m-%d') < '12-21')] = 'fall'

days = pd.Series(np.zeros(len(x)), name='day_of_week')
days[x.dt.day_of_week == 0] = 'Monday'
days[x.dt.day_of_week == 1] = 'Tuesday'
days[x.dt.day_of_week == 2] = 'Wednesday'
days[x.dt.day_of_week == 3] = 'Thursday'
days[x.dt.day_of_week == 4] = 'Friday'
days[x.dt.day_of_week == 5] = 'Saturday'
days[x.dt.day_of_week == 6] = 'Sunday'

year = pd.Series(np.zeros(len(x)), name='year')
year[x.dt.year == 2018] = '2018'
year[x.dt.year == 2019] = '2019'
year[x.dt.year == 2020] = '2020'
year[x.dt.year == 2021] = '2021'

# maint_df
```

**PCA**

```python
from sklearn.decomposition import PCA

# run PCA
pca = PCA(n_components=3)
pca_df = pd.DataFrame(pca.fit_transform(y)).iloc[:,:3] # only keep first 3 dimensions
pca_df.columns = ['PC1', 'PC2', 'PC3']

# group everything together
pca_df = pd.concat([maint_df, days, seasons, year, pca_df], axis=1)
pca_df.head()
```

```python
# plot
sns.lmplot(x='PC1', y='PC2', data=pca_df, fit_reg=False, legend=True,
          hue='maintenance', scatter_kws={'s': 80})
plt.show()
pca.explained_variance_ratio_ # doesn't capture most of data variance
```

**Hierarchial Clustering**

```python
import scipy.cluster.hierarchy as hcluster

# clustering thresholds
thresh = 35
clusters = hcluster.fclusterdata(y, thresh, criterion='distance')

# group everything together
clusters_df = pd.concat([pca_df, pd.Series(clusters, name='clusters')], axis=1)
clusters_df.head()
```

```python
# plot
sns.lmplot(x='PC1', y='PC2', data=clusters_df, fit_reg=False, legend=False,
          hue='clusters', scatter_kws={'s': 80, 'alpha': 1})
plt.show()
```

**Agglomerative Clustering**

```python
from sklearn.cluster import AgglomerativeClustering
aggcluster = AgglomerativeClustering(linkage='ward', n_clusters=3).fit(y)

# group everything together
aggcluster_df = pd.concat([pca_df, pd.Series(aggcluster.labels_, name='aggcluster')], axis=1)
aggcluster_df.head()
```

```python
# plot
sns.lmplot(x='PC1', y='PC2', data=aggcluster_df, fit_reg=False, legend=False,
          hue='aggcluster', scatter_kws={'s': 80, 'alpha': 1})
plt.show()
```

**DBSCAN** density based

```python
from sklearn.cluster import DBSCAN
dbcluster = DBSCAN(eps=30, min_samples=10).fit(y)

# group everything together
dbcluster_df = pd.concat([pca_df, pd.Series(dbcluster.labels_, name='dbcluster')], axis=1)
dbcluster_df.head()
```

```python
# plot
sns.lmplot(x='PC1', y='PC2', data=dbcluster_df, fit_reg=False, legend=False,
          hue='dbcluster', scatter_kws={'s': 80, 'alpha': 0.3})
plt.show()
```

**tSNE**

```python
from sklearn.manifold import TSNE
y_embedded = TSNE(init='random', learning_rate='auto', perplexity=50, n_iter=5000).fit_transform(y) # scaled by sklearn

# group everything together
tSNE_df = pd.concat([maint_df, days, seasons, year, pd.DataFrame(y_embedded, columns=['tSNE1', 'tSNE2'])], axis=1)
tSNE_df.head()
```

## Correlation between segments vs stations

```python
# contribution of first 10 PCs
print(corrm_pca.explained_variance_ratio_.round(2)[:10])
```
